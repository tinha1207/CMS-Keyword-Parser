```python
import pandas as pd
import os

file_name = r"C:\Users\Tin Ha\Downloads\product-search-results 21st April, 2022 08_59 am.xlsx"
df = pd.read_excel(file_name, usecols=["ASIN", "Targeted Search Terms"],keep_default_na=False, na_values=['_',"NaN"])

df
```

```python
map = df.to_dict("index")
map[0]["Targeted Search Terms"]
```

```python
array = []
for v in map.values():
    temp_list = []
    if v["Targeted Search Terms"]:
        temp_list.append(v["Targeted Search Terms"].split(","))
    for x in temp_list:
        for y in x:
            array.append([v["ASIN"],y.strip()])

array
for row in array:
    row[1].replace(r"\xa0", " ")


```

```python
df3 = pd.DataFrame(array,columns=["ASIN","keyword"])


```

```python
df3
```

```python
df4 = df3.pivot_table(index="ASIN",values="keyword",aggfunc="count")
df4.reset_index(inplace=True)
df5 = df[["ASIN"]].merge(df4,how="left")
```

```python
df[["ASIN"]].merge(df4, how="left")
```

```python
df5["keyword"] = df5["keyword"].fillna(0)
```

```python
file_name = "CMS_parsed_keywords.xlsx"

with pd.ExcelWriter(file_name) as writer:
    df3.to_excel(writer,index=False, sheet_name="Parsed Data")
    df5.to_excel(writer,index=False, sheet_name="Report")
    


os.system(f"START EXCEL.EXE {file_name}")
```
